```python
# NAME: gradebook.py

# IMPORT STATEMENTS
# NONE

# FUNCTION DEFINITIONS

# Define  add_students
# ^Create loop for students + grades
def add_students(gradebook):
    while True:
        name = input("Enter student name (or 'done'): ")

        if name == "done":
            break

        grade = float(input(f"Enter {name}'s grade: "))

# Create a dictionary
        gradebook[name] = grade

# PROGRAM LOGIC

# Use dictionary
gradebook = {}

# Add the students
add_students(gradebook)

# Calculate the average grade

# Print report and average
print(gradebook)
```
