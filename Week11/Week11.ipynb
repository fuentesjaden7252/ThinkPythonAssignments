```python
# NAME: gradebook.py

# IMPORT STATEMENTS
# NONE

# FUNCTION DEFINITIONS

# Define  add_students
# ^Create loop for students + grades
def add_students(gradebook):
    while True:
        name = input("Enter student name (or 'done'): ")

        if name == "done":
            break

        grade = float(input(f"Enter {name}'s grade: "))

# Create a dictionary
        gradebook[name] = grade

# Define calculate_average
# ^Create return and calculate average grade of all students
def calculate_average(gradebook):
    if len(gradebook) == 0:
        return 0

    total = sum(gradebook.values())

    average = total / len(gradebook)

    return average

# PROGRAM LOGIC

# Use dictionary
gradebook = {}

# Add the students
add_students(gradebook)

# Calculate the average grade
average = calculate_average(gradebook)

# Print report and average

print("--- Class Report ---", end=" ")

for student, grade in gradebook.items():
    print(f"{student}: {grade}", end=" ")

print(f"Class Average: {average:.2f}")
```
