```python
import json
import os

TASKS_FILE = "tasks.json"
COMPLETED_FILE = "completed.json"

def load_tasks():
    if not os.path.exists(TASKS_FILE):
        return []
    with open(TASKS_FILE, "r") as f:
        return json.load(f)

def save_tasks(tasks):
    with open(TASKS_FILE, "w") as f:
        json.dump(tasks, f, indent=2)

import json
import os

FILE_NAME = "games.json"

# Load existing data if it exists
def load_games():
    if os.path.exists(FILE_NAME):
        with open(FILE_NAME, "r") as file:
            return json.load(file)
    return []

# Save data to file
def save_games(games):
    with open(FILE_NAME, "w") as file:
        json.dump(games, file, indent=4)

# Add a new game
def add_game(games):
    title = input("Game title: ")
    platform = input("Platform (PC, PS5, Xbox, etc.): ")
    genre = input("Genre: ")
    completed = input("Completed? (yes/no): ").lower()

    game = {
        "title": title,
        "platform": platform,
        "genre": genre,
        "completed": completed
    }

    games.append(game)
    print("Game added successfully!\n")

# View all games
def view_games(games):
    if not games:
        print("No games in your library yet.\n")
        return

    print("\n--- Your Game Library ---")
    for i, game in enumerate(games, start=1):
        print(f"{i}. {game['title']} | {game['platform']} | {game['genre']} | Completed: {game['completed']}")
    print()

# Main menu
def main():
    games = load_games()

    while True:
        print("=== GAME LIBRARY TRACKER ===")
        print("1. Add Game")
        print("2. View Games")
        print("3. Save & Exit")

        choice = input("Choose an option: ")

        if choice == "1":
            add_game(games)
        elif choice == "2":
            view_games(games)
        elif choice == "3":
            save_games(games)
            print("Library saved. Goodbye!")
            break
        else:
            print("Invalid choice. Try again.\n")

if __name__ == "__main__":
    main()
```
