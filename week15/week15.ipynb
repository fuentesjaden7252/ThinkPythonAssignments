```python
import json
import os

FILE_NAME = "games.json"

def load_games():
    if not os.path.exists(FILE_NAME):
        return []

    try:
        with open(FILE_NAME, "r") as file:
            data = json.load(file)

            # Ensure data is valid list
            if isinstance(data, list):
                return data
            return []

    except (json.JSONDecodeError, IOError):
        print("Warning: Data file was corrupted. Starting fresh.")
        return []


def save_games(games):
    with open(FILE_NAME, "w") as file:
        json.dump(games, file, indent=4)

def get_non_empty_input(prompt):
    while True:
        value = input(prompt).strip()
        if value:
            return value
        print("Input cannot be empty. Try again.")


def get_yes_no(prompt):
    while True:
        value = input(prompt).strip().lower()
        if value in ["yes", "no"]:
            return value
        print("Please enter 'yes' or 'no'.")

def add_game(games):
    title = get_non_empty_input("Game title: ")
    platform = get_non_empty_input("Platform (PC, PS5, Xbox, etc.): ")
    genre = get_non_empty_input("Genre: ")
    completed = get_yes_no("Completed? (yes/no): ")

    game = {
        "title": title,
        "platform": platform,
        "genre": genre,
        "completed": completed
    }

    games.append(game)
    print("Game added successfully!\n")


def view_games(games):
    if not games:
        print("No games in your library yet.\n")
        return

    print("\n--- Your Game Library ---")
    for i, game in enumerate(games, start=1):
        print(f"{i}. {game['title']} | {game['platform']} | {game['genre']} | Completed: {game['completed']}")
    print()


def completion_rate(games):
    if not games:
        print("No games to analyze.\n")
        return

    completed = sum(1 for g in games if g.get("completed") == "yes")
    total = len(games)

    percent = (completed / total) * 100
    print(f"\nYou have completed {percent:.1f}% of your games.\n")

def what_is_next(games):
    unfinished = [g for g in games if g.get("completed") != "yes"]

    if not games:
        print("No games in your library.\n")
        return

    if not unfinished:
        print("\n🎉 You’ve completed all your games!\n")
        return

    print("\n🎯 What you should play next:")
    print(f"{unfinished[0]['title']} ({unfinished[0]['platform']})\n")

def main():
    games = load_games()

    while True:
        print("=== GAME LIBRARY TRACKER ===")
        print("1. Add Game")
        print("2. View Games")
        print("3. Completion Percentage")
        print("4. What is Next?")
        print("5. Save & Exit")

        choice = input("Choose an option: ").strip()

        if choice == "1":
            add_game(games)
        elif choice == "2":
            view_games(games)
        elif choice == "3":
            completion_rate(games)
        elif choice == "4":
            what_is_next(games)
        elif choice == "5":
            save_games(games)
            print("Library saved. Goodbye!")
            break
        else:
            print("Invalid option. Please choose 1–5.\n")


if __name__ == "__main__":
    main()
```
