```python
# IMPORT STATEMENTS
# (none)

# FUNCTION DEFINITIONS
# 1. Define is_palindrome
#    - Return True
#    - Use string slicing
# 2. Define check_pin
#    - Return True
def is_palindrome(word):
    word = word.lower()
    return word == word[::-1]
    return True
# PROGRAM LOGIC
# 1. Call is_palindrome
# 2. Print final output
# 3. If/else statement
answer = input("Enter a palindrome: ")
if is_palindrome(answer):
    print("Correct! You've pass the test.")
else:
    print("That is wrong.")
```
