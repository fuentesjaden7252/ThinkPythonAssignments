```python
# NAME: rectangle.py
# IMPORT STATEMENTS
# None

# FUNCTION DEFINITIONS
# def calculate_rectanlge
def calculate_rectangle(length, width):
# get length and width
    area = length * width
# calculate area and perimeter
    perimeter = 2 * (length + width)
# return results as tuple
    return (area, perimeter)

# PROGRAM LOGIC
# create input for length and width
# make results into single tuple
# print final results
answer = calculate_rectangle(5, 10)
print(answer)
```
