```python
# NAME: roster.py

# IMPORT STATEMENTS

# None

# FUNCTION DEFINITIONS

# Players added

def add_players(roster):
    while True:
        name = input("Enter player name (or 'quit' to stop): ")

        if name.lower() == "quit":
            break

        roster.append(name)

# Players that are on roster

# PROGRAM LOGIC

# create and empty list

roster = []

# call players added to fill roster

add_players(roster)

# ask for a name to search

# call players on roster

# print

print("Roster players:", roster)
```
