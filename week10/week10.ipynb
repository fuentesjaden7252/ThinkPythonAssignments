```python
# NAME: roster.py

# IMPORT STATEMENTS

# None

# FUNCTION DEFINITIONS

# Players added

def add_players(roster):
    while True:
        name = input("Enter player name (or 'quit' to stop): ")

        if name.lower() == "quit":
            break

        roster.append(name)

# Players that are on roster

def is_on_roster(roster, name):
    for player in roster:
        if player == name:
            return True
    return False

# PROGRAM LOGIC

# create and empty list

roster = []

# call players added to fill roster

add_players(roster)

# ask for a name to search

search_name = input("Who would you like to search for? ")

# call players on roster

# print

if is_on_roster(roster, search_name):
    print("Result: Player found!")
else:
    print("Result: Player not found.")
```
