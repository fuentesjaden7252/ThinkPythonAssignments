```python
# NAME: rentalcar.py
# IMPORT STATEMENTS
# None needed

# FUNCTION DEFINITIONS

def classification_letter(name):
    if name == "B":
        return "Budget"
    if name == "S":
        return "Standard"
    if name == "L":
        return "Luxury"

def classification_cost(name):
    if name == "B":
        return 40
    if name == "S":
        return 60
    if name == "L":
        return 80


def miles_amount(mileage):
    if mileage > 100:
        return mileage - 100
    else:
        return 0


def calculation_for_bill(days, mileage, name):

    daily_cost = classification_cost(name) * days
    charged_miles = miles_amount(mileage)

    mileage_cost = charged_miles * 0.35

    total = daily_cost + mileage_cost
    return total


# PROGRAM LOGIC

name = input("Enter classification code (B,S,L): ").upper()

while name != "B" and name != "S" and name != "L":
    print("Invalid classification code. Please enter B, S, or L.")
    name = input("Enter car classification (B, S, L): ").upper()

start_odometer = int(input("Enter starting odometer reading: "))
end_odometer = int(input("Enter ending odometer reading: "))
days = int(input("Days rented: "))

mileage = end_odometer - start_odometer
charged_miles = miles_amount(mileage)
total_bill = calculation_for_bill(days, mileage, name)


print("-- Rental Summary --")
print("Classification:", classification_letter(name))
print("Total Miles:", mileage)
print("Charged Miles:", charged_miles)
print("Total Bill: $",total_bill)
```
