```python
# Import Statements
# none

# Function definitions

# Put add_expense() function
def add_expense():
# Ask user for category, vendor, amount, and date
  category = input("Enter a category: ")
  vendor = input("Enter vendor: ")

# Use try/except if the user enters letters instead of numbers for amount
  while True:
      try:
          amount = float(input("Enter the amount: "))
          break
      except:
          print("Invalid input. Please enter a number.")

  date = input("Enter the date: ")

# Create dictionary including category, vendor, amount, and date
# Return dictionary
  return {
        "category": category,
        "vendor": vendor,
        "amount": amount,
        "date": date
  }

# Put total_expenses(expenses) Function
# Loop through expenses, then add all amounts together
# Return the total amount
def total_expenses(expenses):
    total = 0
    for exp in expenses:
        total += exp["amount"]
    return total

# Put total_category) (expenses, category name)
# Loop through, if expense matches caategory name then add to total
# Return the total amount
def total_category(expenses, category_name):
    total = 0
    for exp in expenses:
        if exp["category"].lower() == category_name.lower():
            total += exp["amount"]
    return total

# Put total_vendor(expenses, vendor name)
# Loop through, if expense matches vendor name then add to total
# Return the total
def total_vendor(expenses, vendor_name):
    total = 0
    for exp in expenses:
        if exp["vendor"].lower() == vendor_name.lower():
            total += exp["amount"]
    return total

# Program Logic

# Create an empty list called expenses
expenses = []

# Start a loop that continues until user chooses Exit
# Display menu:
# 1.Add New Expense, 2. Get Totals, 3. Exit
while True:
    print("\n--- Expense Tracker ---")
    print("1. Add New Expense")
    print("2. Get Totals")
    print("3. Exit")

    choice = input("Selection: ")

# Selecting 1:
# Call add_expense()
# Add returned dictionary to expenses list
    if choice == "1":
        new_expense = add_expense()
        expenses.append(new_expense)
        print("Expense added successfully!")

# Selecting 2:
# Show totaling options:
# A = Total Expenses, C = Filter by Category, V = Filter by Vendor
    elif choice == "2":
        print("\n--- Totaling Options ---")
        print("A: All Expenses")
        print("C: Filter by Category")
        print("V: Filter by Vendor")

        option = input("Choice: ").upper()

# If A:S
# Call total_all_expenses(expenses)
        if option == "A":
            total = total_expenses(expenses)
            print(f"Total Expenses: ${total:.2f}")

# If C:
# Ask for category name
# Call total_category(expenses, category name)
        elif option == "C":
            category = input("Enter Category to sum: ")
            total = total_category(expenses, category)
            print(f"The total for {category} is: ${total:.2f}")

# If V:
# Ask for vendor name
# Call total_vendor(expenses, vendor name)
# Display totals with 2 decimal places
        elif option == "V":
            vendor = input("Enter Vendor to sum: ")
            total = total_vendor(expenses, vendor)
            print(f"The total for {vendor} is: ${total:.2f}")

# Selecting 3:
    elif choice == "3":
        print("Goodbye!")
        break

# End the program
    else:
        print("Invalid choice. Try again.")
```
