```python
import json
import os

TASKS_FILE = "tasks.json"

# Load tasks
def load_tasks():
    if not os.path.exists(TASKS_FILE):
        return []
    with open(TASKS_FILE, "r") as f:
        return json.load(f)

# Save tasks
def save_tasks(tasks):
    with open(TASKS_FILE, "w") as f:
        json.dump(tasks, f, indent=2)

def add_task(tasks):
    name = input("Task name: ")
    priority = input("Priority (1=High, 2=Medium, 3=Low): ")

    task = {
        "name": name,
        "priority": int(priority)
    }
    tasks.append(task)
    print("Task added!")

def view_tasks(tasks):
    if not tasks:
        print("No tasks found.")
        return

    print("\nYour Tasks:")
    for i, t in enumerate(tasks):
        print(f"{i+1}. {t['name']} (Priority {t['priority']})")

def main():
    tasks = load_tasks()

    while True:
        print("\n--- TASK MANAGER ---")
        print("1. Add Task")
        print("2. View Tasks")
        print("3. Exit")

        choice = input("Choose: ")

        if choice == "1":
            add_task(tasks)
            save_tasks(tasks)
        elif choice == "2":
            view_tasks(tasks)
        elif choice == "3":
            save_tasks(tasks)
            break
        else:
            print("Invalid choice")

if __name__ == "__main__":
    main()
```
