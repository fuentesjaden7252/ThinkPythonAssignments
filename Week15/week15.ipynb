```python
import sys

# Store tasks in a list
tasks = []


def add_task():
    """Add a new task with priority."""
    name = input("Enter task name: ").strip()

    if not name:
        print("Task name cannot be empty.")
        return

    try:
        priority = int(input("Enter priority (1=High, 2=Medium, 3=Low): "))
        if priority not in [1, 2, 3]:
            raise ValueError
    except ValueError:
        print("Invalid priority. Please enter 1, 2, or 3.")
        return

    task = {
        "name": name,
        "priority": priority
    }
    tasks.append(task)
    print(f"Task '{name}' added successfully.\n")


def view_tasks():
    """Display all tasks."""
    if not tasks:
        print("No tasks available.\n")
        return

    print("\nYour Tasks:")
    print("-" * 30)

    # Sort tasks by priority (1 = highest)
    sorted_tasks = sorted(tasks, key=lambda x: x["priority"])

    for idx, task in enumerate(sorted_tasks, start=1):
        priority_map = {1: "High", 2: "Medium", 3: "Low"}
        print(f"{idx}. {task['name']} (Priority: {priority_map[task['priority']]})")

    print()


def main():
    """Main program loop."""
    while True:
        print("=== Personal Task Manager ===")
        print("1. Add Task")
        print("2. View Tasks")
        print("3. Exit")

        choice = input("Choose an option: ").strip()

        if choice == "1":
            add_task()
        elif choice == "2":
            view_tasks()
        elif choice == "3":
            print("Goodbye!")
            sys.exit()
        else:
            print("Invalid choice. Please try again.\n")


if __name__ == "__main__":
    main()

```
