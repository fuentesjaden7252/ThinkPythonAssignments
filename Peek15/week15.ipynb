```python
import json
import os

FILE_NAME = "games.json"

def load_games():
    if os.path.exists(FILE_NAME):
        with open(FILE_NAME, "r") as file:
            return json.load(file)
    return []


def save_games(games):
    with open(FILE_NAME, "w") as file:
        json.dump(games, file, indent=4)

def add_game(games):
    title = input("Game title: ").strip()
    platform = input("Platform (PC, PS5, Xbox, etc.): ").strip()
    genre = input("Genre: ").strip()
    completed = input("Completed? (yes/no): ").strip().lower()

    game = {
        "title": title,
        "platform": platform,
        "genre": genre,
        "completed": completed
    }

    games.append(game)
    print("Game added successfully!\n")


def view_games(games):
    if not games:
        print("No games in your library yet.\n")
        return

    print("\n--- Your Game Library ---")
    for i, game in enumerate(games, start=1):
        print(f"{i}. {game['title']} | {game['platform']} | {game['genre']} | Completed: {game['completed']}")
    print()

def completion_rate(games):
    if not games:
        print("No games to analyze.\n")
        return

    completed = sum(1 for g in games if g["completed"] == "yes")
    total = len(games)

    percent = (completed / total) * 100
    print(f"\nYou have completed {percent:.1f}% of your games.\n")

def what_is_next(games):
    """
    Finds the first game that is NOT completed.
    This acts as a simple 'priority' system.
    """

    for game in games:
        if game["completed"] != "yes":
            print("\n🎯 What you should play next:")
            print(f"{game['title']} ({game['platform']})\n")
            return

    print("\n🎉 You’ve completed all your games!\n")

def main():
    games = load_games()

    while True:
        print("=== GAME LIBRARY TRACKER ===")
        print("1. Add Game")
        print("2. View Games")
        print("3. Completion Percentage")
        print("4. What is Next?")
        print("5. Exit")

        choice = input("Choose an option: ").strip()

        if choice == "1":
            add_game(games)
        elif choice == "2":
            view_games(games)
        elif choice == "3":
            completion_rate(games)
        elif choice == "4":
            what_is_next(games)
        elif choice == "5":
            save_games(games)
            print("Library saved. Goodbye!")
            break
        else:
            print("Invalid choice. Try again.\n")


if __name__ == "__main__":
    main()
```
