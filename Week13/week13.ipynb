```python
# NAME: journal.py

# IMPORT STATEMENTS
# (none)

# FUNCTION DEFINITIONS

# NAME: journal.py

# define write_entry(filename, entry)
# open file in “a” mode
# write entry
# close file
# print saved
def write_entry(filename, entry):
    file = open(filename, 'a')
    file.write(entry + '\n')
    file.close()
    print("Entry saved.")

# define read_journal(filename)
# use try block
# open file in “r” mode
# read contents
# ^print
# close file
# use except FileNotFoundError for except
# print no journal exists
def read_journal(filename):
    try:
        file = open(filename, 'r')
        print("\n--- Your Journal ---")
        contents = file.read()
        print(contents)
        file.close()
    except FileNotFoundError:
        print("No journal entries found yet.")

# PROGRAM LOGIC

# set filename = "journal.txt"
# ask (W)rite, (R)ead, or (Q)ui
# if W ask for entry then call
# if R call read journal
# if Q break loop
# else print invalid input
filename = "journal.txt"
while True:
    choice = input("(W)rite, (R)ead, or (Q)uit? ").lower()
    if choice == 'w':
        entry = input("Enter your entry: ")
        write_entry(filename, entry)
    elif choice == 'r':
        read_journal(filename)
    elif choice == 'q':
        print("Goodbye!")
        break
    else:
        print("Invalid choice. Try again.")
```
